# Spark DataFrame Basic Ops - 2

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/appl_stock.csv', inferSchema=True, header=True ) # Inferschema option is not available with json files
```

```
df.printSchema()
```

```
df.head(3)[0]
```

```
df.filter("Close < 500").select(['Open', 'Close']).show()
```

```
df.filter(df['Close']<500).select(['Open', 'Close', 'Volume']).show()
```

```
# I wanna grab All the rows where closing price less than 200 and opening price greater than 200
df.filter((df['Close']<200) & (df['Open']> 200)).show()
```

```
# WHat date was low price 197?
result = df.filter(df['Low'] == 197.16).collect()
```

```
row = result[0]
```
