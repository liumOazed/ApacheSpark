# Working with DATES

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('Dates').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/appl_stock.csv', inferSchema=True, header=True)
```

```
df.head(1)
```

```
df.select(['Date','Open']).show()
```

```
from pyspark.sql.functions import (dayofmonth,dayofyear, weekofyear,hour,month,
                                  year, format_number, date_format)
```

```
# df.select(month(df['Date'])).show()
```

```
# df.select(year(df['Date'])).show()
new_df = df.withColumn('Year', year(df['Date']))
```

```
result = new_df.groupBy('Year').mean().select(['Year','avg(Close)'])
```

```
new = result.withColumnRenamed('avg(Close)', 'Avg_Close')
```

```
new.select(['Year',format_number('Avg_Close', 2).alias('Avg_Close')]).show()
```
