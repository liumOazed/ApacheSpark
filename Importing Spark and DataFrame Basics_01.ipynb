## Setting up Schema

```
from pyspark.sql.types import StructField,StringType, IntegerType, StructType
```

```
data_schema = [StructField('age', IntegerType(), True),
              StructField('name', StringType(), True)]
```

```
final_struct = StructType(fields=data_schema)
```

```
df = spark.read.json('/FileStore/tables/people.json', schema=final_struct)
```

```
df.printSchema()
```

```
df.show()
```

```
type(df['age'])
```

```
df.select('age').show()
```

```
type(df.head(2)[0])
```

```
df.select(['age','name']).show()
```

```
df.withColumn('double_age', df['age']*2).show()
```

```
df.show()
```

```
df.withColumnRenamed('age', 'my_age').show()
```

```
# Saving as a temp view to query this with sql
df.createOrReplaceGlobalTempView('people')
```

```
results = spark.sql("SELECT * FROM global_temp.people") # This is because temp views are stored in a database named global_temp
```

```
results.show()
```

```
new_results = spark.sql('SELECT * FROM global_temp.people where age = 30')
```

```
new_results.show()
```
